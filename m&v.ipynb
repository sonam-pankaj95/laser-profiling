```python
def cart2pol(x,y,h,k):
    r = np.sqrt((x-h)**2+(y-k)**2)
    t = np.arctan2((y-k),(x-h))
    return(r,t)
```

```python
def pol2cart(rho, phi):
    x = rho * np.cos(phi)
    y = rho * np.sin(phi)
    return(x, y)
```
